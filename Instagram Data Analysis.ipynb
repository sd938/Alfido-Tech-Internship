# Exploratory Data Analysis (EDA)

## Emoji Usage

```python
ax= sns.countplot(x= 'emoji used',data=df,  palette='Set2')

for bars in ax.containers:
    ax.bar_label(bars)
```

* Number of Posts with emoji and that with no emoji are the same and that is 3744

## Timestamp Creation (created Timestamp)

```python
ax= sns.countplot(x= 'created Timestamp',data=df,  palette='Set2')

for bars in ax.containers:
    ax.bar_label(bars)
```

* Timestamp creation on only on 13-04-2023, and at 08:04 is taken into consideration

## Photo ID	

```python
# Top 10 most frequent Photo id values
top_10_photo_ids = df['Photo id'].value_counts().head(10)
print(top_10_photo_ids)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_photo_ids.index, y=top_10_photo_ids.values, palette='Set2')
plt.xlabel('Photo ID')
plt.ylabel('Number of Comments')
plt.title('Top 10 Photo IDs by Number of Comments')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

* There we got photo id no.-247 commented the highest times.

## posted date

```python
ax= sns.countplot(x= 'posted date',data=df,  palette='Set2')

for bars in ax.containers:
    ax.bar_label(bars)
```

* All the posts were done on 14th April.

## Hashtags

```python
ax= sns.countplot(x= 'Hashtags used count',data=df,  palette='Set2')

for bars in ax.containers:
    ax.bar_label(bars)
```

* Maximum people(with a volume of 2496 individuals) who posts use 2 hashtags in there posts.

## Comments

```python
# Top 10 most frequent comments
top_10_comments = df['comment'].value_counts().head(10)
print(top_10_comments)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_comments.index, y=top_10_comments.values, palette='Set2')
plt.xlabel('comments')
plt.ylabel('Number of Comments')
plt.title('Top 10 Photo IDs by Number of Comments')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

* The comment "et et et" comes three times, and that is the most frequent comment in that dataset.
