#  EXPLORATORY DATA ANALYSIS 

## Dish Like

```python
df['dish_liked_split'] = df['dish_liked'].str.split()
df = df.explode('dish_liked_split')
df['dish_liked_split'] = df['dish_liked_split'].str.strip().str.lower()

# Count dish mentions
dish_counts = df['dish_liked_split'].value_counts()

# Get top 10 dishes
top_10_dishes = dish_counts.head(10)

# Display
print("Top 10 Most Sold (Liked) Dishes on Zomato:")
print(top_10_dishes)

top_10_dishes.plot(kind="barh",color="Green")
plt.xlabel("Number of dishes")
plt.ylabel("Dishes")
plt.gca().invert_yaxis()
plt.show()
```

## Cuisines

```python
df['cuisines_split'] = df['cuisines'].str.split()
df = df.explode('cuisines_split')
df['cuisines_split'] = df['cuisines_split'].str.strip().str.lower()

# Count cuisine mentions
cuisines_counts = df['cuisines_split'].value_counts()

# Get top 10 cuisines
top_10_cuisines = cuisines_counts.head(10)

# Display
print("Top 10 Most Sold (Liked) cuisines on Zomato:")
print(top_10_cuisines)

top_10_cuisines.plot(kind="barh",color="maroon")
plt.xlabel("Number of cuisines")
plt.ylabel("cuisines")
plt.gca().invert_yaxis()
plt.show()
```

## Rest Type

```python
# Clean up whitespace and normalize
df['rest_type'] = df['rest_type'].str.strip()

# Count the most common types
top_10_rest_types = df['rest_type'].value_counts().head(10)

# Display top 10 restaurant type(listed_in) on Zomato
print("Top 10 Restaurant Types on Zomato:")
print(top_10_rest_types)

top_10_rest_types.plot(kind='barh', color='purple')
plt.xlabel("number of Restaurent Types")
plt.ylabel("Restaurent Types")
plt.gca().invert_yaxis()
plt.show()
```

## Type of Listing of the Restaurents

```python
# Clean up whitespace and normalize
df['listed_in'] = df['listed_in'].str.strip()

# Count the most common types
top_10_listed_in = df['listed_in'].value_counts().head(10)

# Display top 10 restaurant type(listed_in) on Zomato
print("Top 10 Listed Types on Zomato:")
print(top_10_listed_in)

top_10_listed_in.plot(kind='barh', color='pink')
plt.xlabel("number of Listed Types")
plt.ylabel("Listed Types")
plt.gca().invert_yaxis()
plt.show()
```

## Distribution of Approximate Cost for Two

```python
# Step 1: Remove commas from 'approx_cost' and convert to numeric
df['approx_cost'] = df['approx_cost'].astype(str).str.replace(',', '', regex=False)
df['approx_cost'] = pd.to_numeric(df['approx_cost'], errors='coerce')

# Step 2: Drop rows where 'approx_cost' couldn't be converted
df.dropna(subset=['approx_cost'], inplace=True)

# Step 3: Remove outliers above 99th percentile
threshold = df['approx_cost'].quantile(0.99)
filtered_df = df[df['approx_cost'] <= threshold]

# Step 4: Plot the histogram
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(filtered_df['approx_cost'], bins=30, kde=True, color='orange')
plt.title("Distribution of Approximate Cost for Two (Filtered)")
plt.xlabel("Approximate Cost")
plt.ylabel("Frequency")
plt.show()
```
